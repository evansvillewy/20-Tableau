```python
start_station_date_df = station_df.groupby(['start_date','start station id','start station name','start station latitude',
                                                'start station longitude','gender']).count()[['starttime']]
```

```python
start_station_date_df.reset_index(inplace = True) 
start_station_date_df.dropna()
```

```python
end_station_date_df = station_df.groupby(['start_date','end station id','end station name','end station latitude',
                                                'end station longitude','gender']).count()[['starttime']]
```

```python
end_station_date_df.reset_index(inplace = True) 
end_station_date_df.dropna()
```

```python
start_station_date_df = start_station_date_df.rename(columns={'starttime':'start count',
                                                              'start station id':'station id',
                                                              'start station name':'station name',
                                                              'start station latitude':'latitude',
                                                              'start station longitude':'longitude'})
```

```python
end_station_date_df = end_station_date_df.rename(columns={'starttime':'end count',
                                                          'end station id':'station id',
                                                          'end station name':'station name',
                                                          'end station latitude':'latitude',
                                                          'end station longitude':'longitude'})
```

```python
start_station_date_df['gender'] = start_station_date_df['gender'].map({'0' : 'unknown','1': 'male', '2' : 'female'})
```
